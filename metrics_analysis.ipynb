#### time series baseline

```python
import json
import numpy as np


with open('section7_metrics.json', 'r') as file:
    metrics = json.load(file)

last_items = {}
metrics_stats = {}
for metric, iterations in metrics.items():
    last_items[metric] = [values[-1] for values in iterations.values()]
    
if 'val_precision' in last_items and 'val_recall' in last_items:
    last_items['val_f1_score'] = []
    for precision, recall in zip(last_items['val_precision'], last_items['val_recall']):
        if precision + recall > 0:  # Avoid division by zero
            f1 = 2 * precision * recall / (precision + recall)
        else:
            f1 = 0
        last_items['val_f1_score'].append(f1)



for metric, values in last_items.items():
    metrics_stats[metric] = {
        'mean': np.mean(values),
        'std_dev': np.std(values)
    }

for metric, stats in metrics_stats.items():
    print(f"{metric} - Mean: {stats['mean']*100}, Standard Deviation: {stats['std_dev']*100}")
```

# plots 

```python
import json
import matplotlib.pyplot as plt

plot_metrics = ['val_roc_auc', 'val_pr_auc', 'val_precision', 'val_recall']


def load_metrics(file_name):
    with open(file_name, 'r') as file:
        return json.load(file)

metrics_section7 = load_metrics('section7_metrics.json')
metrics_section8_word2vec = load_metrics('section8_metrics_word2vec.json')
metrics_section8_fasttext = load_metrics('section8_metrics_fasttext.json')
metrics_section8_concat = load_metrics('section8_metrics_concat.json')
metrics_section9_word2vec = load_metrics('section9_metrics_word2vec.json')
metrics_section9_fasttext = load_metrics('section9_metrics_fasttext.json')
metrics_section9_concat = load_metrics('section9_metrics_concat.json')

focus_iteration = '3'

data_for_plot = {}
for metric in plot_metrics:
    data_for_plot[metric] = {
        'Time series baseline': metrics_section7[metric].get(focus_iteration, None),
        'multimodal baseline_Word2Vec': metrics_section8_word2vec[metric].get(focus_iteration, None),
        'multimodal baseline_FastText': metrics_section8_fasttext[metric].get(focus_iteration, None),
        'multimodal baseline_Concat': metrics_section8_concat[metric].get(focus_iteration, None),
        'proposed_Word2Vec': metrics_section9_word2vec[metric].get(focus_iteration, None),
        'proposed_FastText': metrics_section9_fasttext[metric].get(focus_iteration, None),
        'proposed_Concat': metrics_section9_concat[metric].get(focus_iteration, None)
    }
    
def calculate_f1(precision, recall):

    if isinstance(precision, list) and isinstance(recall, list):
     
        if len(precision) != len(recall):
            raise ValueError("Precision and Recall lists must be of the same length")

        f1_scores = []
        for p, r in zip(precision, recall):
            if (p + r) == 0:
                f1_scores.append(0)  # Avoid division by zero
            else:
                f1_scores.append(2 * (p * r) / (p + r))
        return f1_scores

    if (precision + recall) == 0:
        return 0  
    return 2 * (precision * recall) / (precision + recall)


f1_scores = {}
for model in data_for_plot['val_precision']:
    precision = data_for_plot['val_precision'][model]
    recall = data_for_plot['val_recall'][model]
    f1_score = calculate_f1(precision, recall)
    f1_scores[model] = f1_score


data_for_plot['f1_score'] = f1_scores


for metric, values in data_for_plot.items():
    plt.figure(figsize=(8, 6))  
    for model, value in values.items():
        if value is not None: 
            plt.plot(value, label=model,linewidth= 4)
    plt.title(f'{metric} Comparison')
    plt.xlabel('Epochs')
    plt.ylabel(metric)
    plt.legend()
    plt.show()
```
